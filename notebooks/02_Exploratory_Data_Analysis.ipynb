# 02_Exploratory_Data_Analysis
## Analisis Pendapatan Kepala Keluarga RT 013 / RW 004 Dusun Ciranggon

---
## 1. Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (10, 5)

print('Library berhasil diimport')
```

---
## 2. Load Clean Dataset

```python
df = pd.read_csv('../data/pendapatan_rt013_clean.csv')

print(f'Dataset terload: {df.shape[0]} baris, {df.shape[1]} kolom')
df.head()
```

---
## 3. Statistik Umum

```python
print('Statistik Umum Dataset')
print('=' * 40)
print(f'Jumlah Kepala Keluarga    : {len(df)}')
print(f'Total Pendapatan          : Rp{df["pendapatan"].sum():,}')
print(f'Rata-rata Pendapatan      : Rp{df["pendapatan"].mean():,.0f}')
print(f'Median Pendapatan         : Rp{df["pendapatan"].median():,.0f}')
print(f'Pendapatan Tertinggi      : Rp{df["pendapatan"].max():,}')
print(f'Pendapatan Terendah       : Rp{df["pendapatan"].min():,}')
print(f'Rata-rata Umur            : {df["umur"].mean():.0f} tahun')
print(f'Rata-rata Jml Tanggungan  : {df["jumlah_tanggungan"].mean():.1f}')
print(f'Rata-rata Lama Bekerja    : {df["lama_bekerja"].mean():.0f} tahun')
```

---
## 4. Distribusi Pekerjaan

```python
pekerjaan_counts = df['pekerjaan'].value_counts()

print('Distribusi Pekerjaan:')
print(pekerjaan_counts)
print()

fig, ax = plt.subplots()
bars = ax.bar(pekerjaan_counts.index, pekerjaan_counts.values, 
              color=['#4C72B0', '#55A868', '#DD8452'])
ax.set_title('Distribusi Pekerjaan Kepala Keluarga', fontsize=14, fontweight='bold')
ax.set_xlabel('Pekerjaan')
ax.set_ylabel('Jumlah KK')
ax.bar_label(bars, padding=2)
plt.tight_layout()
plt.savefig('../charts/distribusi_pekerjaan.png', dpi=120, bbox_inches='tight', facecolor='white')
plt.show()
```

---
## 5. Distribusi Pendidikan

```python
pendidikan_order = ['Tidak Sekolah', 'SD/Sederajat', 'SMP/Sederajat', 'SMA/Sederajat']
pendidikan_counts = df['pendidikan'].value_counts().reindex(pendidikan_order).dropna()

print('Distribusi Pendidikan:')
print(pendidikan_counts)
print()

fig, ax = plt.subplots()
colors_p = ['#DD8452', '#55A868', '#4C72B0', '#C44E52']
bars = ax.bar(pendidikan_counts.index, pendidikan_counts.values, color=colors_p)
ax.set_title('Distribusi Pendidikan Kepala Keluarga', fontsize=14, fontweight='bold')
ax.set_xlabel('Pendidikan')
ax.set_ylabel('Jumlah KK')
ax.bar_label(bars, padding=2)
plt.tight_layout()
plt.savefig('../charts/distribusi_pendidikan.png', dpi=120, bbox_inches='tight', facecolor='white')
plt.show()
```

---
## 6. Distribusi Pendapatan

```python
print('Statistik Pendapatan:')
print(df['pendapatan'].describe())
print()

fig, ax = plt.subplots()
ax.hist(df['pendapatan'] / 1_000_000, bins=10, edgecolor='white', 
        color='#4C72B0')
ax.set_title('Distribusi Pendapatan Kepala Keluarga', fontsize=14, fontweight='bold')
ax.set_xlabel('Pendapatan (juta Rupiah)')
ax.set_ylabel('Jumlah KK')
plt.tight_layout()
plt.savefig('../charts/histogram_pendapatan.png', dpi=120, bbox_inches='tight', facecolor='white')
plt.show()
```

---
## 7. Pendapatan Berdasarkan Pekerjaan

```python
pekerjaan_stats = df.groupby('pekerjaan')['pendapatan'].agg(['mean', 'median', 'min', 'max'])
pekerjaan_stats = pekerjaan_stats.round(0).astype(int)

print('Pendapatan per Pekerjaan:')
print(pekerjaan_stats)
print()

fig, ax = plt.subplots()
pekerjaan_order = df.groupby('pekerjaan')['pendapatan'].median().sort_values().index
bp = ax.boxplot([df[df['pekerjaan'] == p]['pendapatan'] / 1_000_000 for p in pekerjaan_order],
                tick_labels=pekerjaan_order, patch_artist=True)
colors = ['#DD8452', '#55A868', '#4C72B0']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
ax.set_title('Pendapatan Berdasarkan Pekerjaan', fontsize=14, fontweight='bold')
ax.set_xlabel('Pekerjaan')
ax.set_ylabel('Pendapatan (juta Rupiah)')
plt.tight_layout()
plt.savefig('../charts/pendapatan_per_pekerjaan.png', dpi=120, bbox_inches='tight', facecolor='white')
plt.show()
```

---
## 8. Pendapatan Berdasarkan Pendidikan

```python
pendidikan_stats = df.groupby('pendidikan')['pendapatan'].agg(['mean', 'median', 'min', 'max'])
pendidikan_stats = pendidikan_stats.round(0).astype(int)

print('Pendapatan per Pendidikan:')
print(pendidikan_stats)
print()

fig, ax = plt.subplots()
pendidikan_order = ['Tidak Sekolah', 'SD/Sederajat', 'SMP/Sederajat', 'SMA/Sederajat']
available = [p for p in pendidikan_order if p in df['pendidikan'].unique()]
bp = ax.boxplot([df[df['pendidikan'] == p]['pendapatan'] / 1_000_000 for p in available],
                tick_labels=available, patch_artist=True)
colors_p2 = ['#DD8452', '#55A868', '#4C72B0', '#C44E52']
for patch, color in zip(bp['boxes'], colors_p2):
    patch.set_facecolor(color)
ax.set_title('Pendapatan Berdasarkan Pendidikan', fontsize=14, fontweight='bold')
ax.set_xlabel('Pendidikan')
ax.set_ylabel('Pendapatan (juta Rupiah)')
plt.tight_layout()
plt.savefig('../charts/korelasi.png', dpi=120, bbox_inches='tight', facecolor='white')
plt.show()
```

---
## 9. Korelasi
Matriks korelasi antar variabel numerik.

```python
kolom_numerik = ['umur', 'lama_bekerja', 'pendapatan', 'jumlah_tanggungan']
korelasi = df[kolom_numerik].corr()

print('Matriks Korelasi:')
print(korelasi.round(3))
print()

fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(korelasi.values, cmap='RdBu_r', vmin=-1, vmax=1)

ax.set_xticks(range(len(kolom_numerik)))
ax.set_yticks(range(len(kolom_numerik)))
ax.set_xticklabels(kolom_numerik, rotation=45, ha='right')
ax.set_yticklabels(kolom_numerik)

for i in range(len(kolom_numerik)):
    for j in range(len(kolom_numerik)):
        ax.text(j, i, f'{korelasi.values[i, j]:.2f}', 
                ha='center', va='center', fontsize=10,
                color='white' if abs(korelasi.values[i, j]) > 0.5 else 'black')

ax.set_title('Matriks Korelasi', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
