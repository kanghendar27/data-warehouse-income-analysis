## 4. Transform
Membersihkan dan memastikan dataset siap digunakan.

```python
# 4a. Pastikan tipe data sesuai
df['id_kk'] = df['id_kk'].astype(int)
df['umur'] = df['umur'].astype(int)
df['lama_bekerja'] = df['lama_bekerja'].astype(int)
df['pendapatan'] = df['pendapatan'].astype(int)
df['jumlah_tanggungan'] = df['jumlah_tanggungan'].astype(int)
df[['nama_kk', 'jenis_kelamin', 'pendidikan', 'pekerjaan', 'status_rumah', 'kendaraan', 'bantuan_sosial', 'rt', 'rw', 'dusun', 'desa', 'kecamatan', 'kabupaten']] = df[['nama_kk', 'jenis_kelamin', 'pendidikan', 'pekerjaan', 'status_rumah', 'kendaraan', 'bantuan_sosial', 'rt', 'rw', 'dusun', 'desa', 'kecamatan', 'kabupaten']].astype(str)

print("Tipe data telah disesuaikan")
df.dtypes
```

```python
# 4b. Validasi nilai kosong
null_count = df.isnull().sum().sum()
print(f"Total nilai kosong: {null_count}")

if null_count == 0:
    print("Tidak ada nilai kosong. Dataset siap digunakan.")
else:
    print("Terdapat nilai kosong. Perlu penanganan lanjutan.")
```

```python
# 4c. Validasi duplikasi
dup_count = df.duplicated().sum()
print(f"Jumlah data duplikat: {dup_count}")

if dup_count == 0:
    print("Tidak ada duplikasi. Dataset siap digunakan.")
else:
    df = df.drop_duplicates()
    print(f"Duplikasi telah dihapus. Tersisa {df.shape[0]} baris.")
```

```python
# 4d. Format kolom string (trim whitespace)
string_cols = ['nama_kk', 'jenis_kelamin', 'pendidikan', 'pekerjaan', 'status_rumah', 'kendaraan', 'bantuan_sosial']
for col in string_cols:
    df[col] = df[col].str.strip()

print("Kolom string telah diformat (whitespace dihapus)")
```

```python
# Tampilkan hasil akhir setelah transformasi
df.head()
```

---
## 5. Load
Menyimpan dataset bersih ke file CSV.

```python
df.to_csv('../data/pendapatan_rt013_clean.csv', index=False)

print("Dataset bersih berhasil disimpan ke data/pendapatan_rt013_clean.csv")
```

---
## 6. Ringkasan ETL

```python
print("=" * 40)
print("        RINGKASAN ETL")
print("=" * 40)
print(f"\nJumlah data          : {df.shape[0]} baris")
print(f"Jumlah kolom         : {df.shape[1]} kolom")
print(f"Jumlah data kosong   : {null_count}")
print(f"Jumlah duplikasi     : {dup_count}")
print(f"\nKesimpulan:")

if null_count == 0 and dup_count == 0:
    print("Dataset bersih dan siap digunakan untuk tahap analisis selanjutnya.")
elif null_count > 0 and dup_count == 0:
    print("Dataset memiliki nilai kosong. Perlu dibersihkan sebelum analisis.")
elif null_count == 0 and dup_count > 0:
    print("Dataset memiliki duplikasi. Duplikasi telah dihapus.")
else:
    print("Dataset memiliki nilai kosong dan duplikasi. Keduanya telah ditangani.")
print("=" * 40)
```
